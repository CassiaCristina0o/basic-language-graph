# library imports


```python
import pandas as pd 
import seaborn as sns

tmdb = pd.read_csv('tmdb_5000_movies.csv')
```

# English language data extraction compared to other languages 

```python
language_count = tmdb['original_language'].value_counts().to_frame().reset_index()
```

```python
total_language = tmdb['original_language'].value_counts()
```

```python
overall_total = total_language.sum ()
```

```python
total_en = total_language.loc['en']
```

```python
total_other_languages = overall_total - total_en 
```

```python
information = { 'languages' : ['en','other'],
         'total' : [total_en  , total_other_languages]}
```

```python
information = pd.DataFrame (information)
```

```python
sns.barplot(x = "languages",y = "total",data=information)
```

# Extraction of data with other languages only

```python
total_other_languages_2 = tmdb.query('original_language != "en"').original_language.value_counts()
```

```python
total_languages_except_en = tmdb.query('original_language != "en"')
```

```python
sns.catplot(x="original_language",kind='count', data=total_languages_except_en, aspect=2,palette="dark:#5A9_r",order=total_other_languages_2.index)
```
